```python
import tensorflow as tf
import numpy as np
```

```python
#1. Declare two constant tensors that have the values of 15 and 45. Add these two tensors and print out the results.
import tensorflow as t
a = t.constant([15])
b = t.constant([45])
res = t.math.add(a,b)
sess = tf.Session()
print(sess.run(res))
```

```python
#2. Declare two variable tensors, a and b, that are initialized with scalar values of 2.75 and 8.5. Find their product and print out the result.
a = tf.Variable([2.75])
b = tf.Variable([8.5])
c = tf.math.multiply(a,b)
init = tf.global_variables_initializer()
sess = tf.Session()
#run variables initializer
sess.run(init)
print(sess.run(c))
```

```python
#3. Create two placeholders: x and y - that are both scalars of 32-bit floats. Assign 5.25 to x and 12.6 to y, multiply them together, and print out the results.
x = tf.placeholder(tf.float32, shape=())
y = tf.placeholder(tf.float32, shape=())
z = tf.math.multiply(x,y)
sess = tf.Session()
print(sess.run(z, feed_dict={x: 5.25, y:12.6}))
```

```python
#4 Create one placeholder: z - that is an N-Dimensional array (N can be >= 1) that can have any shape (shape = None). Feed this vector [1, 3, 5, 7, 9] into z and multiply it by 3.
# Display the results. 
z = tf.placeholder(tf.int32, shape=[5])
res = tf.math.multiply(z,3)
print(sess.run(res, feed_dict={z: [1, 3, 5, 7, 9]}))
```

```python
#5. Create a constant tensor that is a matrix of the shape (8, 8). The matrix is initialized with all ones (1). Create a variable tensor that is also a matrix of the shape (8, 8) and initialized with random integer values between 0 and 99. Add these two tensors and display the results. 

x1 = np.ones([8,8])
x = tf.constant(x1)
print(" X is ", sess.run([x]))

y = np.zeros([8,8])
for i in range(8):
  for j in range(8):
    y[i][j] = np.random.randint(1,98)
y = tf.Variable(y)
init = tf.global_variables_initializer()
sess.run(init)
print("\n Y is", sess.run([y]))

res = tf.math.add(x,y)
init = tf.global_variables_initializer()
sess.run(init)
print('\n Addittion Result =', sess.run([res]))
```
